```python
#mean/ median user tenure :
get_date_posts_per_user = df_posts.groupby('OwnerUserId').apply(lambda x: sorted(x['CreationDate'].tolist()))
```

```python
def cal_no_of_days(list_dates_posts):
    days = []
    #list_dates_posts = list_dates_posts
    if(len(list_dates_posts) == 1):
        return (1,1)
    else:
        start = pd.to_datetime(list_dates_posts[0])
        #print(start)i = 1
        for i in range(len(list_dates_posts)):
            #print(list_dates_posts[i])
            diff = pd.to_datetime(list_dates_posts[i]) - start
            #print(diff)
            if(diff.days==0 and i!=0):
                days.append(1)
            else:
                days.append(diff.days)
        #print(days)
        avg_user_tenure = sum(days)/(len(days)-1)
        median_user_tenure = statistics.median(days)
        return (avg_user_tenure,median_user_tenure)
        
```

```python
user_tenure = []
for index, row in get_date_posts_per_user_df.iterrows():
    list_dates_posts = (row["Date"])
    #print(list_dates_posts)
    res = cal_no_of_days(list_dates_posts)
    #print(res)
    avg_user_tenure = res[0]
    median_tenure = res[1]
    user_tenure.append((row["OwnerUserId"],avg_user_tenure,median_tenure))
    
    
    
    
```

```python
user_tenure_df = pd.DataFrame(user_tenure)
user_tenure_df.columns =['OwnerUserId', 'AvgUserTenure', 'MedianUserTenure']
```
